```python
productAPrice = 20
productBPrice = 40
productCPrice = 50

giftWrapFeePerUnit = 1
shippingFeePerPackage = 5
maxUnitsPerPackage = 10

cart = {}

totalQuantity = 0
totalCost = 0
discountApplied = None

def applyFlatDiscount(totalCost):
    if totalCost >= 200:
        return 10
    return 0

def applyBulkDiscount(item):
    if item["quantity"] > 10:
        return (item["quantity"] * item["price"] * 0.05)
    return 0

def applyBulkDiscountTotal(totalQuantity, totalCost):
    if totalQuantity > 20:
        return (totalCost * 0.10)
    return 0

def applyTieredDiscount(cart):
    eligible_products = [product for product, item in cart.items() if item["quantity"] > 15]

    if totalQuantity > 30 and eligible_products:
        discount = 0
        highest_discount_product = None
        max_discount = 0

        for product, item in cart.items():
            if product in eligible_products:
                product_discount = (item["quantity"] - 15) * item["price"] * 0.50
                if product_discount > max_discount:
                    max_discount = product_discount
                    highest_discount_product = product

        if highest_discount_product:
            discount = max_discount
            return discount

    return 0

productAQuantity = int(input(f"Product A quantity : "))
productAGiftWrap = input("wrap Product A as a gift? (y/n): ").strip().lower() == "y"
cart["Product A"] = {"quantity": productAQuantity, "price": productAPrice, "giftWrap": productAGiftWrap}

productBQuantity = int(input(f"Product B quantity : "))
productBGiftWrap = input("wrap Product B as a gift? (y/n): ").strip().lower() == "y"
cart["Product B"] = {"quantity": productBQuantity, "price": productBPrice, "giftWrap": productBGiftWrap}

productCQuantity = int(input(f"Product C quantity : "))
productCGiftWrap = input("wrap Product C as a gift? (y/n): ").strip().lower() == "y"
cart["Product C"] = {"quantity": productCQuantity, "price": productCPrice, "giftWrap": productCGiftWrap}

for product, item in cart.items():
    totalQuantity += item["quantity"]
    totalCost += item["quantity"] * item["price"]
    totalCost -= applyBulkDiscount(item)

totalCost -= applyFlatDiscount(totalCost)
tieredDiscount = applyTieredDiscount(cart)

if tieredDiscount > 0:
    discountApplied = "Tiered 50% Discount"
    totalCost -= tieredDiscount

shippingFee = (totalQuantity // maxUnitsPerPackage) * shippingFeePerPackage

giftWrapFeeTotal = sum(item["quantity"] for item in cart.values() if item["giftWrap"]) * giftWrapFeePerUnit

totalCost += shippingFee + giftWrapFeeTotal

print("\nCart Details:")
for product, item in cart.items():
    print(f"{product}: Quantity: {item['quantity']}, Total Cost: ${item['quantity'] * item['price']}")

print(f"\nSubtotal: ${totalCost}")
if discountApplied:
    print(f"{discountApplied} Applied: -${tieredDiscount}")
print(f"Shipping Fee: ${shippingFee}")
print(f"Gift Wrap Fee: ${giftWrapFeeTotal}")
print(f"Total: ${totalCost}")
```
